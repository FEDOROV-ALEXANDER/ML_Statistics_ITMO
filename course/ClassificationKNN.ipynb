```python
import numpy as np
import matplotlib.pyplot as plt

# Создаем более детальную сетку
x = np.linspace(0, 24, 1000)
y = np.linspace(0, 24, 1000)
X, Y = np.meshgrid(x, y)

# Условие: -2 < y - x < 1
condition = (-2 < Y - X) & (Y - X < 1)

# Создаем изображение
plt.figure(figsize=(10, 10))
plt.imshow(condition, extent=[0, 24, 0, 24], origin='lower', 
           cmap='Blues', alpha=0.7, aspect='equal')

# Добавляем граничные линии
x_line = np.linspace(0, 24, 100)
plt.plot(x_line, x_line - 2, 'red', linewidth=3, label='y = x - 2')
plt.plot(x_line, x_line + 1, 'green', linewidth=3, label='y = x + 1')

# Настройки графика
plt.xlabel('X', fontsize=12)
plt.ylabel('Y', fontsize=12)
plt.xlim(0, 24)
plt.ylim(0, 24)
ax = plt.gca()
ax.set_aspect('equal', adjustable='box')
plt.title(' -2 < y - x < 1', fontsize=14)
plt.grid(True, alpha=0.3)
# plt.legend()
# plt.colorbar(label='Выполнение условия')

plt.show()
```
