# Напоминание: производная и дифференциал


## Основные понятия

Пусть $U, V$ — нормированные пространства, $f: U \to V$. Функция $f$ называется дифференцируемой в точке $x_0$, внутренней для $U$, если

$$
f(x_0 + h) - f(x_0) = f'(x_0)[h] + o(\|h\|), \quad h \in U, \ h \to 0,
$$

а $f'(x_0)$ линейно по $h$. Линейный оператор $f'(x_0)$ называется **производной** $f$ в точке $x_0$. Величина $f'(x_0)[h] = f'(x_0)\,dx$ называется **дифференциалом** $f$ в точке $x_0$ и обозначается $df = df(x_0)[h]$.

Написанное равенство следует понимать как

$$
\frac{\|f(x_0 + h) - f(x_0) - f'(x_0)[h]\|_V}{\|h\|_U} \xrightarrow[h \to 0]{} 0.
$$

Мы будем рассматривать лишь конечномерные пространства, поэтому конкретный выбор нормы не важен — все нормы эквивалентны.

Напомним, что будем понимать под градиентом.

Пусть $f: U \to V$.

* Если $U, V \subset \mathbb{R}$, то градиентом $f$ назовём

$$
\nabla f(x) = f'(x).
$$

* Если $U \subset \mathbb{R}^n$, $V \subset \mathbb{R}$, то градиентом $f$ назовём

$$
\nabla f(x) =
\begin{pmatrix}
\dfrac{\partial f}{\partial x_1} \\
\vdots \\
\dfrac{\partial f}{\partial x_n}
\end{pmatrix}.
$$

* Если $U \subset \mathbb{R}^{n \times m}$, $V \subset \mathbb{R}$, то градиентом $f$ назовём матрицу

$$
\nabla f(X) =
\begin{pmatrix}
\dfrac{\partial f}{\partial x_{11}} & \cdots & \dfrac{\partial f}{\partial x_{1m}} \\
\vdots & \ddots & \vdots \\
\dfrac{\partial f}{\partial x_{n1}} & \cdots & \dfrac{\partial f}{\partial x_{nm}}
\end{pmatrix}.
$$

* Если $U \subset \mathbb{R}$, $V \subset \mathbb{R}^p$, то градиентом $f$ назовём

$$
\nabla f(x) =
\begin{pmatrix}
f_1'(x) \\
\vdots \\
f_p'(x)
\end{pmatrix}.
$$

* Если $U \subset \mathbb{R}^n$, $V \subset \mathbb{R}^p$, то градиентом $f$ назовём якобиан

$$
\nabla f(x) = J(x) =
\begin{pmatrix}
\dfrac{\partial f_1}{\partial x_1} & \cdots & \dfrac{\partial f_1}{\partial x_n} \\
\vdots & \ddots & \vdots \\
\dfrac{\partial f_p}{\partial x_1} & \cdots & \dfrac{\partial f_p}{\partial x_n}
\end{pmatrix}.
$$

Как градиент связан с дифференциалом? Нам же, для поиска подозрительных на экстремум точек, нужно занулять градиент.

\begin{array}{c|c|c|c}
U \backslash V & \mathbb{R} & \mathbb{R}^p & \mathbb{R}^{p \times q} \\ \hline
\mathbb{R} & f'(x)\,dx & \nabla f(x)\,dx & ? \\
\mathbb{R}^n & \nabla f(x)^\top dx & J(x)\,dx & ? \\
\mathbb{R}^{n \times m} & \langle \nabla f, dX \rangle = \operatorname{Tr}(\nabla f(x)^\top dX) & ? & ?
\end{array}

Там, где стоят вопросительные знаки, появляются тензоры; такие случаи нас мало интересуют. Вообще, нам интересен первый столбец (в основном).

## Полезности (примеры)

### Пример 1

Пусть $f(x) = c^\top x$, где $c, x \in \mathbb{R}^n$. Покажем, что

$$
df[h] = c^\top h.
$$

Это моментально следует из того, что

$$
c^\top (x + h) - c^\top x = c^\top h.
$$

Считая, что градиент — это столбец, получаем

$$
\nabla f = c.
$$

В дифференциальной записи это означает, что

$$
d(c^\top x) = c^\top dx.
$$

### Пример 2

Пусть $f(x) = x^\top c$. Так как

$$
(x + h)^\top c - x^\top c = h^\top c = c^\top h,
$$

то $df[h] = c^\top h$, и, следовательно,

$$
\nabla f = c.
$$

Значит

$$
d(x^\top c) = c^\top dx
$$

### Пример 3

Покажем, что для $f(x) = A x$ справедливо $df[h] = A h$. Действительно,

$$
A(x + h) - A x = A h,
$$

откуда

$$
\nabla f = A.
$$

В дифференциальной форме получим

$$
d(Ax) = Adx
$$

**Примечание:** нужно обратить внимание на размерности пространств откуда и куда действует функция и понять, что из себя будет представлять градиент (якобиан)

### Пример 4

Найдём дифференциал функции $f(x) = x^\top A x$. Выполним преобразования:

$$
(x + h)^\top A (x + h) - x^\top A x =
$$

$$
= x^\top A h + h^\top A x + h^\top A h =
$$

$$
= x^\top (A + A^\top) h + h^\top A h.
$$

Первое слагаемое линейно по $h$, а второе — $o(\|h\|)$, поскольку $\|h^\top A h\| \leq \|A\| \|h\|^2$.

Следовательно,

$$
df(x)[h] = x^\top (A + A^\top) h,
$$

а так как градиент — столбец, то

$$
\nabla f = (A + A^\top) x.
$$

В случае симметричной матрицы $A$ получаем $\nabla f = 2 A x$.

В дифференциальной форме получим

$$
d(x^\top A x) = x^\top(A+A^\top)dx = ((A+A^\top)x)^\top dx
$$

## Правила дифференцирования

Конечно, кроме таблицы стандартных производных, полезно отметить и некоторые правила дифференцирования (прямо как в курсе анализа).

Пусть $U, V$ — нормированные пространства, $x$ — внутренняя точка $U$. Справедливы следующие утверждения:
*  Если $f$ — постоянное отображение, то $df = 0$.

*  Если $f \equiv \mathrm{id}$, то $f' = 1$ и $df = dx$.

*  Пусть $f, g: U \to V$. Если $f, g$ дифференцируемы в $x$, $\alpha, \beta \in \mathbb{R}$, то

$$
d(\alpha f + \beta g) = \alpha\, df + \beta\, dg.
$$

*  Пусть $\varphi: U \to \mathbb{R}$, $f: U \to V$. Если $\varphi, f$ дифференцируемы в $x$, то

$$
d(\varphi f) = d\varphi \cdot f + \varphi \cdot df.
$$

*  Пусть $\varphi: U \to \mathbb{R}$, $f: U \to V$, $\varphi(x) \ne 0$. Тогда

$$
d\!\left(\frac{f}{\varphi}\right) = \frac{\varphi\, df - d\varphi \cdot f}{\varphi^2}.
$$

*  Пусть $f: U \to V$, $g: V \to W$, $V, W \subset \mathbb{R}$. Если $f$ дифференцируемо в $x$, а $g$ — в $f(x)$, то

$$
d(g(f(x))[h] = dg(f(x))[df(x)[h]].
$$

*  Пусть $f: U \to \mathbb{R}^{n \times m}$, $g: U \to \mathbb{R}^{m \times k}$. Если $f, g$ дифференцируемы в $x$, то

$$
d(fg)(x)[h] = f(x)\, dg(x)[h] + (df(x)[h]) g(x).
$$

### Правила дифференцирования для матриц

Для матриц эти правила принимают следующий вид.
Пусть $A, B$ — фиксированные матрицы. Тогда (в случае, когда объекты определены):
*  $dA = 0$,
*  $d(AXB) = A\, dX\, B$,
*  $d(\alpha X) = \alpha\, dX$,
*  $d(XY) = X\, dY + dX\, Y$.

Пример:

Покажем, что
$$
dX^{-1} = -X^{-1}\, dX\, X^{-1}.
$$
Будем считать, что $X^{-1}$ дифференцируема. Так как $X X^{-1} = I$, то
$$
dX \cdot X^{-1} + X \cdot dX^{-1} = 0 \quad \Rightarrow \quad dX^{-1} = -X^{-1}\, dX\, X^{-1}.
$$

# Задача 1.1

$$
f(x) = \ln(x^TAx).
$$

Вычислить $df$, записать градиент

### Решение

Найдем дифференциал первого порядка.

$$
df = d(\ln(x^TAx)) = \frac{1}{x^TAx}d(x^TAx).
$$

Заметим, что

$$
d(x^TAx) = x^T(A+A^T)dx,
$$

тогда

$$
df = \frac{x^T(A+A^T)}{x^TAx}dx.
$$

Запишем градиент из соотношения $df = \nabla f(x)^\top dx$. Для этого с учетом того, что $x^\top A x$ — число получим

$$
df = \left( \frac{(A+A^T)x}{x^TAx} \right)^\top dx,
$$

откуда

$$
\nabla f(x) = \frac{(A+A^T)x}{x^TAx}
$$

# Задача 2.1

$$
f(x) = ||x||_2^3
$$

Вычислить $df$, записать градиент

### Решение

$$
df = d(x^Tx)^{\frac{3}{2}} = \frac{3}{2}(x^Tx)^{\frac{1}{2}}\cdot d(x^Tx) = \frac{3}{2}(x^Tx)^{\frac{1}{2}}\cdot 2x^Tdx.
$$

Тогда:

$$
df = 3(x^Tx)^{\frac{1}{2}}x^Tdx = 3||x||_2x^Tdx,
$$

значит

$$
\nabla f(x) =  3||x||_2x
$$

# Задача 3.1

$$
f(\theta) = \ln\left(1+e^{(\theta, x)}\right).
$$

Вычислить $df$, записать градиент

### Решение

$$
df = d\left(\ln\left(1+e^{(\theta, x)}\right)\right) = \frac{1}{1+e^{(\theta, x)}}\cdot e^{(\theta, x)} \cdot d(\theta, x) =
$$

$$
= \frac{e^{(\theta, x)}}{1+e^{(\theta, x)}} \cdot d\left(x^T\theta\right) = \frac{e^{(\theta, x)}}{1+e^{(\theta, x)}} \cdot x^T d\theta.
$$

Тогда

$$
df = \left(\frac{e^{(\theta, x)}}{1+e^{(\theta, x)}} \cdot x\right)^\top d\theta
$$
и
$$
\nabla f(\theta) = \frac{e^{(\theta, x)}}{1+e^{(\theta, x)}} \cdot x
$$

# Задача 4

$$
f(\theta) = ||X\theta-Y||_2^2,
$$

$X$ — матрица размера $[n\times p]$,

$\theta$ — вектор размера $[p\times 1]$,

$Y$ — вектор размера $[n\times 1]$.

Вычислить $df$.

Вычислить $\theta$, если $\nabla f = 0$.

### Решение

1. Найдем дифференциал

$$
df = d\left(\left(X\theta-Y\right)^T\left(X\theta-Y\right)\right).
$$

$$
df = d\left( (\theta^T X^T - Y^T)(X\theta -Y)\right) =
$$

$$
= d \left(\theta^T X^T X\theta - \theta^T X^T Y - Y^T X\theta + Y^T Y   \right).
$$

Теперь можно дифференцировать каждое слагаемое. Начнем со второго, первое рассмотрим последним.

* $d(\theta^T X^T Y)$

Здесь $X^T Y$ — постоянный вектор (не зависит от $\theta)$, значит

$$
d(\theta^T X^T Y) = Y^T X d\theta.
$$

* $d(Y^T X\theta)$

Аналогично предыдущему, получаем

$$
d(Y^T X\theta) = Y^T X d\theta
$$

* $d(Y^T Y) = 0$,

Так как $Y^T Y$ — константа.

* $d \left(\theta^T X^T X\theta \right)$.

Тут можно применить правило дифференциала произведения (для матриц), либо заметить, что это квадратичная форма относительно матрицы $X^T X$, значит ее дифференциал будет иметь вид $\theta^T (X^T X + (X^T X)^T)d\theta$. Тогда для этого слагаемого можем записать

$$
d \left(\theta^T X^T X\theta \right) = 2\theta^T X^T Xd\theta.
$$

Собирая все вместе, получим

$$
df = 2\theta^T X^T Xd\theta - 2 Y^T X d\theta =
$$

$$
=2(\theta^T X^T - Y^T)Xd\theta = \left(2X^T(X\theta - Y)\right)^Td\theta
$$

2. Получим требуемое выражение для $\theta$ при $\nabla f = 0$.

$$
2X^T(X\theta - Y) = 0,
$$

откуда

$$
X^TX\theta = X^TY,
$$

и

$$
\theta = \left(X^TX\right)^{-1}\cdot X^TY
$$

#### Примечание

Можно было в самом начале не раскрывать скобки, а рассмотреть $d\left(\left(X\theta-Y\right)^T\left(X\theta-Y\right)\right)$ как дифференциал произведения по правилам для матриц:

$$
d(XY) = X\, dY + dX\, Y.
$$

Тогда

$$
d\left(X\theta-Y\right)^T\left(X\theta-Y\right) + \left(X\theta-Y\right)^Td \left(X\theta-Y\right)=
$$

$$
=d\left(\theta^TX^T\right)\left(X\theta-Y\right) + \left(X\theta-Y\right)^T\cdot Xd\theta =
$$

$$
=d\theta^TX^T\left(X\theta-Y\right) + \left(X\theta-Y\right)^T\cdot Xd\theta.
$$

Рассмотрим первое слагаемое. Выпишем размерности: $[1\times p][p\times n][n\times 1]$, итого — скаляр, который можно транспонировать, значит

$$
df = \left(X\theta-Y\right)^T\left(d\theta^TX^T\right)^T + \left(X\theta-Y\right)^T\cdot Xd\theta =
$$

$$
= \left(X\theta-Y\right)^T\cdot Xd\theta + \left(X\theta-Y\right)^T\cdot Xd\theta = 2\left(X\theta-Y\right)^T\cdot Xd\theta =
$$

$$
 = (2X^T(X\theta - Y))^Td\theta
$$

# Задача 5* (якобиан)

$$
f(x) = x x^\top x,
$$

Вычислить $df$, записать $\nabla f$.

**Дисклеймер:** функция действует из векторов в векторы, поэтому в данном случае $\nabla f$ — якобиан.

### Решение

$$
df = d(xx^\top x) = d((xx^\top)x).
$$

В таком случае под знаком дифференциала у нас произведение матрицы $xx^\top$ на вектор $x$. Применяя (дважды) правило производной произведения (для матриц), получим

$$
df = xx^\top dx + d(xx^\top)x = xx^\top dx + (xdx^\top + (dx)x^\top)x =
$$

$$
= xx^\top dx + xdx^\top x + dx x^\top x.
$$

Пользуясь ассоциативностью перепишем в виде

$$
df = xx^\top dx + x(dx^\top x) + dx (x^\top x).
$$

С учетом того, что в скобках стоят скаляры, получим

$$
df = xx^\top dx + xx^\top dx + x^\top x dx = \left(2xx^\top + x^\top x I\right) dx,
$$

где $I$ — единичная матрица.

Значит якобиан будет иметь вид

$$
2xx^\top + x^\top x I
$$

**Замечание:**

Можно рассмотреть случай $df = d(x(x^\top x))$, тогда имеем дифференциал произведения вектора $x$ на скаляр $x^\top x$, а значит можем применять правила дифференцирования для скалярных функций. Тогда

$$
df = x d(x^\top x) + x^\top x dx.
$$

Учитывая, что $d(x^\top x) = 2^\top dx$, получим

$$
df = 2xx^\top dx + x^\top x dx = \left( 2xx^\top + x^\top x I \right)dx.
$$


# Задача 6

Найти $df$ и $d^2f$ функции, записать градиент и гессиан

$$
f(x) = \frac{1}{2}x^\top Ax - b^\top x + c,
$$
где

$x$ — вектор размерности $n$,

$A$ — симметричная матрица размера $[n\times n]$,

$b$ — постоянный вектор размерности $n$,

$c$ — число.

**Напоминание:** в случае

$$
d^2f = dx_1^\top H dx
$$

матрица $H$ называется гессианом. Гессиан может также обозначаться как $\nabla^2 f(x)$.

### Решение

1. Вычислим дифференциал функции по правилам дифференцирования:
$$
df = d\left( \frac{1}{2} x^\top A x - b^\top x + c \right) = \frac{1}{2} d(x^\top A x) - d(b^\top x) + dc.
$$

Поскольку $ c $ — константа, $ dc = 0 $.  

По условию $ A^\top = A $, поэтому:
$$
d(x^\top A x) = x^\top (A + A) dx = 2 x^\top A dx.
$$

Подставляем:
$$
df = \frac{1}{2} \cdot 2 x^\top A dx - b^\top dx = x^\top A dx - b^\top dx = (A x - b)^\top dx.
$$

Следовательно, градиент равен:
$$
\boxed{\nabla f(x) = A x - b}.
$$

2. Второй дифференциал $ d^2f $

Теперь найдём $ d^2f = d(df) $. Имеем:
$$
df = (A x - b)^\top dx.
$$

При вычислении $ d^2f $ вектор $ dx $ считается постоянным (не зависит от $ x $), поэтому обозначим его $ dx_1 $. Тогда:
$$
d^2f = d\left( (A x - b)^\top dx_1 \right) = d\left( (A x - b)^\top \right) \cdot dx_1.
$$

Вычислим дифференциал:
$$
d\left( (A x - b)^\top \right) = d(x^\top A^\top - b^\top) = d(x^\top A) = (dx)^\top A,
$$
поскольку $ A^\top = A $ и $ b $ — константа.

Подставляем:
$$
d^2f = dx^\top A dx_1 = dx_1^\top A dx.
$$

Следовательно, гессиан равен:
$$
\boxed{H = \nabla^2 f(x) = A}.
$$



# Задача 7

Найти $df$ и $d^2f$ функции, записать градиент и гессиан

$$
f(x) = ||x||_2,
$$

где $ x \in \mathbb{R}^n $ — вектор.

### Решение

Запишем функцию в виде:
$$
f(x) = (x^\top x)^{1/2}.
$$

Вычислим дифференциал, используя правило дифференцирования сложной функции:
$$
df = d\left( (x^\top x)^{1/2} \right) = \frac{1}{2} (x^\top x)^{-1/2} \cdot d(x^\top x).
$$

При этом
$$
d(x^\top x) = x^\top dx + dx^\top x = 2x^\top dx.
$$

Подставляем:
$$
df = \frac{1}{2} (x^\top x)^{-1/2} \cdot 2x^\top dx = \frac{x^\top dx}{\sqrt{x^\top x}}.
$$

Заметим, что $ \sqrt{x^\top x} = \|x\|_2 $, поэтому:
$$
df = \frac{x^\top}{\|x\|_2} dx.
$$

Это выражение имеет вид $ df = (\nabla f)^\top dx $, следовательно, градиент равен:
$$
\boxed{\nabla f(x) = \frac{x}{\|x\|_2}}.
$$

**Замечание:** Это выражение определено только при $ x \ne 0 $. При $ x = 0 $ функция $ f(x) = \|x\|_2 $ не дифференцируема — в нуле она имеет «острие» (как модуль в одномерном случае).

**Второй дифференциал**

Теперь найдём $ d^2f = d(df) $. Имеем:
$$
df = \left( \frac{x}{\|x\|_2} \right)^\top dx.
$$

При вычислении $ d^2f $ вектор $ dx $ считается постоянным (не зависит от $ x $). Обозначим его $ dx_1 $.

Тогда:
$$
df = \left( \frac{x}{\|x\|_2} \right)^\top dx_1.
$$

Дифференцируем это выражение по $ x $:
$$
d^2f = d\left( \left( \frac{x}{\|x\|_2} \right)^\top dx_1 \right) = \left[ d\left( \frac{x}{\|x\|_2} \right) \right]^\top dx_1.
$$

Вычислим дифференциал вектора $ \frac{x}{\|x\|_2} $. Обозначим $ r = \|x\|_2 = \sqrt{x^\top x} $, тогда:
$$
d\left( \frac{x}{r} \right) = \frac{ r \cdot dx - x \cdot dr }{r^2}.
$$

Найдём $ dr $:
$$
dr = d(\|x\|_2) = df = \frac{x^\top dx}{r}.
$$

Подставляем:
$$
d\left( \frac{x}{r} \right) = \frac{ r \cdot dx - x \cdot \left( \frac{x^\top dx}{r} \right) }{r^2} = \frac{1}{r} dx - \frac{1}{r^3} x (x^\top dx).
$$

Заметим, что $ x (x^\top dx) = (x x^\top) dx $ — это матрица $ x x^\top $, умноженная на вектор $ dx $.

Таким образом:
$$
d\left( \frac{x}{r} \right) = \left( \frac{1}{r} I - \frac{1}{r^3} x x^\top \right) dx,
$$

где $I$ — единичная (симметричная) матрица, а $xx^\top$ — тоже симметричная матрица, поэтому их линейная комбинация — тоже симметричная матрица.

Тогда:
$$
d^2f = \left[ \left( \frac{1}{r} I - \frac{1}{r^3} x x^\top \right) dx \right]^\top dx_1 = dx^\top \left( \frac{1}{r} I - \frac{1}{r^3} x x^\top \right) dx_1.
$$

Итак, получили:
$$
\boxed{d^2f = dx^\top \left( \frac{1}{\|x\|_2} I - \frac{1}{\|x\|_2^3} x x^\top \right) dx_1}.
$$

Согласно определению, если
$$
d^2f = dx^\top H \, dx_1,
$$
то матрица $ H $ называется гессианом $ \nabla^2 f(x) $.

Следовательно:
$$
\boxed{\nabla^2 f(x) = \frac{1}{\|x\|_2} I - \frac{1}{\|x\|_2^3} x x^\top}.
$$

**Примечание:**
- Гессиан симметричен, так как $ x x^\top $ — симметричная матрица.  
- При $ x \ne 0 $ гессиан определён.  
- При $ x = 0 $ гессиан не существует — функция не дважды дифференцируема в нуле.  


# Задача 1.2

В условиях задачи 1, то есть для

$$
f(x) = \ln(x^TAx)
$$

и

$$
df = \frac{x^T(A+A^T)}{x^TAx}dx.
$$

вычислить $d^2f$, записать гессиан.

### Решение

$$
d^2f = d(d(f)).
$$

Теперь $dx$ — некоторый постоянный вектор. Обозначим его $dx_1$, тогда

$$
d^2f = d\left(\frac{x^T(A+A^T)dx_1}{x^TAx}\right).
$$

Напомним, что

$$
d\left(\frac{f}{\varphi}\right) = \frac{\varphi df - d\varphi f}{\varphi^2},
$$

значит

$$
d^2f = \frac{x^TAx \cdot d(x^T(A+A^T)dx_1)-d(x^TAx) \cdot x^T(A+A^T)dx_1}{(x^TAx)^2}.
$$

Выражение, стоящее под знаком дифференциала в первом слагаемом является скаляром (можно проверить по размерностям), поэтому его можно транспонировать, учитывая, что

$$
(A\cdot B)^T = B^T\cdot A^T,
$$

получим, что

$$
d^2f = \frac{x^TAx \cdot d(dx_1^T(A+A^T)x)-x^T(A+A^T)dx \cdot x^T(A+A^T)dx_1}{(x^TAx)^2}=
$$

$$
= \frac{x^TAx \cdot dx_1^T(A+A^T)dx-x^T(A+A^T)dx \cdot x^T(A+A^T)dx_1}{(x^TAx)^2}.
$$

Выражения $x^T(A+A^T)dx$ и $x^T(A+A^T)dx_1$ — скаляры, воспользовавшись ассоциативностью, имеем

$$
d^2f = \frac{x^TAx \cdot dx_1^T(A+A^T)dx- x^T(A+A^T)dx_1 \cdot x^T(A+A^T)dx}{(x^TAx)^2}=
$$

$$
 = dx_1^T\frac{A+A^T}{x^TAx}dx - dx_1^T\frac{(A+A^T)x \cdot x^T(A+A^T)}{(x^TAx)^2}dx=
$$

$$
 = dx_1^T \left(\frac{A+A^T}{x^TAx} - \frac{(A+A^T)x \cdot x^T(A+A^T)}{(x^TAx)^2}\right)dx.
$$

Откуда

$$
\nabla^2f = \frac{A+A^T}{x^TAx} - \frac{(A+A^T)x \cdot x^T(A+A^T)}{(x^TAx)^2}
$$

# Задача 2.2

В условиях задачи 2, то есть для

$$
f(x) = ||x||_2^3
$$

и

$$
df = 3||x||_2x^Tdx,
$$

вычислить $d^2f$, записать гессиан.

### Решение

Найдем вторую производную, заменив $dx$ на $dx_1$.

$$
d^2f = d(3||x||_2x^Tdx) = d(3||x||_2)x^Tdx_1 + 3||x||_2\cdot d(x^Tdx_1) =
$$

$$
= 3d\left((x^Tx)^{\frac{1}{2}}\right)\cdot x^Tdx_1 + 3||x||_2 \left(dx^Tdx_1 + x^Td(dx_1)\right).
$$

Заметим, что $dx^Tdx_1$ — скаляр, который можно транспонировать, а $d(dx_1) = 0$. Тогда

$$
d^2f = \frac{3}{2}||x||_2^{-1}d(x^Tx) \cdot x^Tdx_1 + 3||x||_2dx_1^Tdx =
$$

$$
= 3||x||_2^{-1}x^Tdx\cdot x^Tdx_1 + 3||x||_2dx_1^Tdx =
$$

$$
=3||x||_2^{-1}dx_1^Tx\cdot x^Tdx + 3||x||_2dx_1^Tdx.

d^2f = dx_1^T3\left(||x||_2^{-1}x\cdot x^T + ||x||_2I\right)dx.

Откуда

$$
\nabla^2 f(x) = 3\left(||x||_2^{-1}x\cdot x^T + ||x||_2I\right).
